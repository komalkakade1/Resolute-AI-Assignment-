```python
data=data.drop('location.1', axis=1)
```

```python
data.dropna(inplace=True)
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
inside_activities = data[data['position'].str.lower() == 'inside']
outside_activities = data[data['position'].str.lower() == 'outside']
```

```python
inside_activities
```

```python
data['time'] = data['time'].astype(str)
```

```python
data['time'] = pd.to_datetime(data['time'])
```

```python
duration_data = data.groupby(['date', 'position'])['time'].apply(lambda x: (x.max() - x.min()).total_seconds() / 60).reset_index(name='total_duration')
```

```python
duration_data
```

```python
activity_counts = data.groupby(['date', 'activity']).size().reset_index(name='activity_count')
```

```python
activity_counts
```

```python
picking_placing_counts = activity_counts[activity_counts['activity'].isin(['picked', 'placed'])]
```

```python
picking_placing_counts_pivot = picking_placing_counts.pivot(index='date', columns='activity', values='activity_count').fillna(0)
```

```python
picking_placing_counts_pivot.reset_index(inplace=True)
```

```python
picking_placing_counts_pivot.columns.name = None  # Remove column name
picking_placing_counts_pivot.columns = ['date', 'placing_count', 'picking_count']
```
